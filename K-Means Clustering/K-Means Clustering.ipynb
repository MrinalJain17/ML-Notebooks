```python
import numpy as np
import matplotlib.pyplot as plt
from cluster import KMeans

%matplotlib inline

data_points = np.loadtxt('sample_2D_points.csv', dtype='int', delimiter=',')
```

```python
n_clusters = 2
cluster = KMeans(data=data_points, k=n_clusters)
cluster.fit()
predictions = cluster.predict(data_points)
for i, center in enumerate(cluster.cluster_centers):
    print('Cluster {}: {}'.format(i, center))

temp = np.array(list(zip(data_points, predictions)))
plt.scatter(cluster.cluster_centers[:,0], cluster.cluster_centers[:,1], marker='x')
for i in range(n_clusters):
    index = np.where(temp[:,1] == i)
    plt.scatter(data_points[index][:,0], data_points[index][:,1])
```
